```python
from matplotlib import pyplot as plt
import torch
from positional_embeddings import PositionalEmbedding


emb = PositionalEmbedding(128, 'learnable', scale=25)

xs = torch.linspace(-2, 2, 100)

embeddings = emb(xs)

xs.shape
embeddings.shape

plt.imshow(embeddings.detach().cpu().T)
plt.xlabel('Position')
plt.ylabel('Embedding Dimension')
```
